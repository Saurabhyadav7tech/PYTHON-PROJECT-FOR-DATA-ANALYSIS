# Converting Date Columns:

```python
data["Order Date"] = pd.to_datetime(data["Order Date"])
data["Ship Date"] = pd.to_datetime(data["Ship Date"])
```

```python
data.info()
```

```python
data["Order Month"] = data["Order Date"].dt.month
data[" Order Year"] = data["Order Date"].dt.year
data["Month_name"] = data["Order Date"].dt.strftime("%b")
data["Order WeekNo"] = data["Order Date"].dt.dayofweek
```

```python
data.head(3)
```

# Monthly Sales Analysis:

```python
sales_by_month = data.groupby("Month_name").agg({"Sales":"sum"})
print(sales_by_month)
```

```python
grouped = data.groupby("Order Month")["Sales"].sum().reset_index()
print(grouped)
```

```python
sns.set(rc = {"figure.figsize":(12,4)})
ax = sns.lineplot(grouped, x = "Order Month" , y = "Sales" , marker = "d"   )
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("MONTHLY SALES ANALYSIS" , fontsize = 12)
plt.show()
```

The chart shows monthly sales with significant fluctuations, including a notable spike in March and November.  
Sales generally increase toward the end of the year, peaking in November and slightly declining in December.

# Sales Analysis By Category:

```python
sales_by_category = data.groupby("Category").agg({"Sales":"sum"})
print(sales_by_category)
```

```python
plt.figure(figsize = (5,4))
plt.pie(sales_by_category["Sales"],labels = sales_by_category.index  , autopct = "%.2f" , shadow = True , colors = ["brown","green","orange"])
plt.title("SALES ANALYSIS BY CATEGORY",fontsize = 12)
plt.show()
```

The Chart shows that Technology contributes the highest sales percentage at 36.4%, followed by Furniture at 32.3%.
Where as Office Supplies account for the lowest sales percentage, contributing 31.3%.

# Sales Analysis By Sub-Category:

```python
sales_by_sub_category = data.groupby("Sub-Category").agg({"Sales":"sum"}).sort_values(by = "Sales" , ascending = False).head(10)
print(sales_by_sub_category)
```

```python
sns.set(rc = {"figure.figsize":(17,4)})
ax = sns.barplot(sales_by_sub_category , x = "Sub-Category" , y = "Sales" , palette = "viridis")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("SALES ANALYSIS BY SUB-CATEGORY" , fontsize = 12)
plt.show()
```

The chart shows that Phones and Chairs lead in sales, while Appliances, Bookcases, and Copiers have the lowest performance among sub-categories

# Monthly Profit Analysis:

```python
monthly_profit = data.groupby("Order Month").agg({"Profit":"sum"})
print(monthly_profit)
```

```python
sns.set(rc = {"figure.figsize":(17,4)})
sns.lineplot(monthly_profit , x = "Order Month",y = "Profit" , marker = "d" )
plt.title("MONTHLY PROFIT" , fontsize = 12)
plt.show()
```

The chart shows a steady overall increase in monthly profits throughout the year, with fluctuations in certain months, 
ending with the highest profit in December.

```python
profit_category = data.groupby("Category").agg({"Profit":"sum"})
print(profit_category)
```

```python
plt.figure(figsize = (5,4))
ex = [0.2,0,0]
plt.pie(profit_category["Profit"] ,labels = profit_category.index , autopct = "%.2f" , shadow = True  , explode = ex)
plt.title("PROFIT BY CATEGORY",fontsize = 12)
plt.show()
```

# Profit By Sub-Category:

```python
profit_subcategory = data.groupby("Sub-Category").agg({"Profit":"sum"})
print(profit_subcategory)
```

```python
sns.set(rc = {"figure.figsize":(17,5)})
ax = sns.barplot(profit_subcategory , x= "Sub-Category" , y = "Profit" , color = "brown")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("PROFIT BY SUB-CATEGORY",fontsize = 12)
plt.show()
```

The chart shows that the Copiers and Phones are the most profitable sub-categories,
while Tables and Bookcases incur the highest losses,indicating potential areas for strategy optimization.

# Sales And Profit By Customer Segments:

```python
sales_profit_by_segments = data.groupby("Segment").agg({"Sales":"sum","Profit":"sum"}).reset_index()
print(sales_profit_by_segments)
```

```python
sns.set(rc = {"figure.figsize":(8,4)})
sales_profit_by_segments.plot(kind='bar', x='Segment', y=['Sales', 'Profit'])
plt.title("SALES AND PROFIT BY CUSTOMER SEGMENTS" , fontsize = 12)
plt.show()
```

The chart shows that the majority of the sales and profit come from the consumer and then corporate , while Home office occur the less sales and profit
